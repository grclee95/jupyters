```python
import pandas as pd
import numpy as np

df = pd.read_csv('./data/auto-mpg.csv', header=None)
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 
              'acceleration', 'model year', 'origin', 'name']
# print(df.head())
df.info()

#horsepower는 마력인데 마력은 숫자데이터입니다.
#현재 자료형을 보면 object로 만들어져 있습니다.
```

```python
#horsepower 컬럼의 자료형을 float으로 변경해서 대입 - 에러 발생
#df['horsepower'] = df['horsepower'].astype('float')

#중복된 데이터를 제외하고 출력 - ? 라는 값이 존재
#print(df['horsepower'].unique())

#?라는 값을 가진 데이터를 어떻게 할 것인지 결정
#?값을 None으로 설정하고 None인 데이터를 제거
df['horsepower'].replace('?', np.nan, inplace=True)
df.dropna(subset=['horsepower'], axis=0, inplace=True)
df['horsepower']=df['horsepower'].astype('float')
```

```python
df['origin'].replace({1:'미국', 2:'유럽', 3:'일본'}, inplace=True)
df['origin'] = df['origin'].astype('category') #범주형으로 변환
df.info()
df.corr()
```
